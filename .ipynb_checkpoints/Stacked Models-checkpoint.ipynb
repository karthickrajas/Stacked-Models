#### Target variable analysis

```python
#function for nomality test

def normality_test(x):
    shapiro_stat,shapiro_pvalue = shapiro(x)
    anderson_stat,anderson_pvalue,anderson_critical = anderson(x)
    kstest_stat,kstest_pvalue = kstest(x,cdf ='norm')
    data = {'shapiro': [shapiro_stat, shapiro_pvalue , 'Nan'], 'anderson': [anderson_stat, anderson_pvalue, anderson_critical],'kstest':[kstest_stat,kstest_pvalue,"Nan"]}
    normality_df = pd.DataFrame.from_dict(data, orient='index', columns=['statistic','P-value','critical_values'])
    return normality_df
```

```python
#function for fitting a normal curve for examination
def feature_analysis(x,name):
    sns.distplot(x , fit=norm)
    # Get the fitted parameters used by the function
    (mu, sigma) = norm.fit(x)
    print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

    #Now plot the distribution
    plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],loc='best')
    plt.ylabel('Frequency')
    plt.title(name)

    #Get also the QQ-plot
    fig = plt.figure()
    res = stats.probplot(x, plot=plt)
    plt.show()
    
    #Results from the normality tests
    print(normality_test(train.SalePrice))
    
```

```python
feature_analysis(train.SalePrice,'SalePrice')
```

#### Univariate plots

```python
def univariate_plots(df):
    #Divding categorical and numerical variable
    numerical_columns = list(df.columns[df.dtypes=="int64"])
    numerical_columns = numerical_columns + list(df.columns[df.dtypes=="float64"])
    categorical_columns = list(df.columns[df.dtypes=="object"])
    print("Univariate plot on Numerical variables ., number of variables {}".format(len(numerical_columns)))
    df[numerical_columns].hist(figsize=(16,20), bins=50, xlabelsize=8, ylabelsize=8)
    plt.show()
    print("Univariate plot on Categorical variables ., number of variables {}".format(len(categorical_columns)))
    fig, axes = plt.subplots(round(len(categorical_columns) / 3), 3, figsize=(24,36))
    for i, ax in enumerate(fig.axes):
        if i < len(categorical_columns):
            ax.set_xticklabels(ax.xaxis.get_majorticklabels(), rotation=90)
            sns.countplot(x=df[categorical_columns[i]], alpha=0.7, data=df, ax=ax)
    fig.tight_layout()
    plt.show()
```

#### Bivariate plot

```python
def bivariate_plot(df,name):
    #Divding categorical and numerical variable
    numerical_columns = list(df.columns[df.dtypes=="int64"])
    numerical_columns = numerical_columns + list(df.columns[df.dtypes=="float64"])
    categorical_columns = list(df.columns[df.dtypes=="object"])
    fig, ax = plt.subplots(round(len(numerical_columns) / 3), 3, figsize=(24,36))
    for i, ax in enumerate(fig.axes):
        if i <= len(numerical_columns) - 1:
            sns.regplot(x=df[numerical_columns[i]],y=df[name], ax=ax)
    fig.tight_layout()
    plt.show()
    
    
    fig, ax = plt.subplots(round(len(categorical_columns) / 1), 1, figsize=(12, 30))
    for i, ax in enumerate(fig.axes):
        if i <= len(categorical_columns) - 1:
            sns.boxplot(x=df[categorical_columns[i]],y=df[name], ax=ax)
    fig.tight_layout()
    plt.show()
```

```python
bivariate_plot(train,'SalePrice')
```

#### Filling in NA values

```python
all_data["PoolQC"] = all_data["PoolQC"].fillna("None")
all_data["Alley"] = all_data["Alley"].fillna("None")
all_data["Fence"] = all_data["Fence"].fillna("None")
all_data["FireplaceQu"] = all_data["FireplaceQu"].fillna("None")
all_data["LotFrontage"] = all_data.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
for col in ('GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'):
    all_data[col] = all_data[col].fillna('None')
for col in ('GarageYrBlt', 'GarageArea', 'GarageCars'):
    all_data[col] = all_data[col].fillna(0)
for col in ('BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath'):
    all_data[col] = all_data[col].fillna(0)
for col in ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'):
    all_data[col] = all_data[col].fillna('None')
all_data["MasVnrType"] = all_data["MasVnrType"].fillna("None")
all_data["MasVnrArea"] = all_data["MasVnrArea"].fillna(0)
all_data['MSZoning'] = all_data['MSZoning'].fillna(all_data['MSZoning'].mode()[0])
all_data = all_data.drop(['Utilities'], axis=1)
all_data["Functional"] = all_data["Functional"].fillna("Typ")
all_data['Electrical'] = all_data['Electrical'].fillna(all_data['Electrical'].mode()[0])
all_data['KitchenQual'] = all_data['KitchenQual'].fillna(all_data['KitchenQual'].mode()[0])
all_data['Exterior1st'] = all_data['Exterior1st'].fillna(all_data['Exterior1st'].mode()[0])
all_data['Exterior2nd'] = all_data['Exterior2nd'].fillna(all_data['Exterior2nd'].mode()[0])
all_data['SaleType'] = all_data['SaleType'].fillna(all_data['SaleType'].mode()[0])
all_data['MSSubClass'] = all_data['MSSubClass'].fillna("None")
all_data["MiscFeature"] = all_data["MiscFeature"].fillna("None")
```

#### checking for skewness

```python
numeric_feats = all_data.dtypes[all_data.dtypes != "object"].index

# Check the skew of all numerical features
skewed_feats = all_data[numeric_feats].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
print("\nSkew in numerical features: \n")
skewness = pd.DataFrame({'Skew' :skewed_feats})
skewness.head(10)
```

#### Boxcox transformation

##### Boxcox transformation of 1+x
##### Note that setting  Î»=0  is equivalent to log1p 

```python
skewness = skewness[abs(skewness) > 0.75]
print("There are {} skewed numerical features to Box Cox transform".format(skewness.shape[0]))

from scipy.special import boxcox1p
skewed_features = skewness.index
lam = 0.15
for feat in skewed_features:
    #all_data[feat] += 1
    all_data[feat] = boxcox1p(all_data[feat], lam)
```
